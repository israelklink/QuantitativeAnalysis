# Análise de fundo de investimento

Coletando os dados da CVM

```python
import pandas as pd
pd.set_option("display.max_colwidth", 150) #A largura máxima da coluna para 150
```

Coletando informações no site da CVM
- https://dados.cvm.gov.br/dados/FI/DOC/INF_DIARIO/DADOS/
- https://dados.cvm.gov.br/dados/FI/CAD/DADOS/

A análise sera feita no ano/mês

```python
# definindo funções para coleta de dados

def busca_informes_cvm(ano, mes):
  url = 'http://dados.cvm.gov.br/dados/FI/DOC/INF_DIARIO/DADOS/inf_diario_fi_{:02d}{:02d}.zip'.format(ano,mes)
  return pd.read_csv(url, sep=';')
```

```python
def busca_cadastro_cvm(ano, mes, dia):
  url = 'http://dados.cvm.gov.br/dados/FI/CAD/DADOS/cad_fi.csv'.format(ano, mes, dia)
  return pd.read_csv(url, sep=';', encoding='ISO-8859-1')
```

>Buscando dados no site da CVM

```python
informes_diarios = busca_informes_cvm(2023,4)  # anos de 2023, mês abril
```

```python
informes_diarios
```

```python
cadastro_cvm = busca_cadastro_cvm(2023,5,1)
```

```python
cadastro_cvm
```

>Definindo filtros para os Fundos de Investimento


```python
minimo_cotistas = 100
```

> Manipulando os dados e aplicando filtros
- informes_diarios[informes_diarios['NR_COTST'] >= minimo_cotistas]: Filtra as linhas do DataFrame informes_diarios onde o valor na coluna 'NR_COTST' é maior ou igual a minimo_cotistas;
- .pivot(index='DT_COMPTC', columns='CNPJ_FUNDO', values=['VL_TOTAL', 'VL_QUOTA', 'VL_PATRIM_LIQ', 'CAPTC_DIA', 'RESG_DIA']): Realiza uma operação de pivô no DataFrame resultante. Ele redefine o DataFrame, tendo 'DT_COMPTC' como índice, 'CNPJ_FUNDO' como colunas e os valores das colunas 'VL_TOTAL', 'VL_QUOTA', 'VL_PATRIM_LIQ', 'CAPTC_DIA' e 'RESG_DIA' como os dados.  a saída será um novo DataFrame onde as datas são índices.


```python
fundos = informes_diarios[informes_diarios['NR_COTST'] >= minimo_cotistas].pivot(index='DT_COMPTC', columns='CNPJ_FUNDO', values=['VL_TOTAL',	'VL_QUOTA',	'VL_PATRIM_LIQ',	'CAPTC_DIA',	'RESG_DIA'])
```

```python
fundos
```

>Normalizando os dados de cotas para efeitos comparativos

```python
cotas_normalizadas = fundos['VL_QUOTA'] / fundos['VL_QUOTA'].iloc[0]
```

```python
cotas_normalizadas
```

###Fundos de Investimento com os melhores desempenhos em Abril de 2023

```python
melhores = pd.DataFrame()
melhores['retorno(%)'] = (cotas_normalizadas.iloc[-1].sort_values(ascending=False)[:5] - 1) * 100
melhores
```

>Buscando dados dos Fundos de Investimento pelo CNPJ

```python
for cnpj in melhores.index:
  fundo = cadastro_cvm[cadastro_cvm['CNPJ_FUNDO'] == cnpj]
  melhores.at[cnpj, 'Fundo de Investimento'] = fundo['DENOM_SOCIAL'].values[0]
  melhores.at[cnpj, 'Classe'] = fundo['CLASSE'].values[0]
  melhores.at[cnpj, 'PL'] = fundo['VL_PATRIM_LIQ'].values[0]
```

```python
melhores
```

###Fundos de Investimento com os piores desempenhos em Abril de 2023

```python
piores = pd.DataFrame()
piores['retorno(%)'] = (cotas_normalizadas.iloc[-1].sort_values(ascending=True)[:5] - 1) * 100
piores
```

>Buscando dados dos Fundos de Investimento pelo CNPJ

```python
for cnpj in piores.index:
  fundo = cadastro_cvm[cadastro_cvm['CNPJ_FUNDO'] == cnpj]
  piores.at[cnpj, 'Fundo de Investimento'] = fundo['DENOM_SOCIAL'].values[0]
  piores.at[cnpj, 'Classe'] = fundo['CLASSE'].values[0]
  piores.at[cnpj, 'PL'] = fundo['VL_PATRIM_LIQ'].values[0]
```
