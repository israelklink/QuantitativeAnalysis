```python
data=pd.read_excel('data.xlsx')  # leitura dos dados em excel

if 'Unnamed: 0' in data.columns:
    data.drop('Unnamed: 0', axis=1, inplace=True)
else:
    print("'Unnamed: 0' column not found in the DataFrame.")
```

```python
data['symbol']
```

```python
# start = '2021-08-01'
# end = '2021-08-30'

tickers=data['symbol']

# ['BTC-USD', 'ETH-USD', 'HEX-USD','USDT-USD','ADA-USD','BNB-USD','XRP-USD','SOL-USD','USDC-USD','DOGE-USD',
#         'UNI-USD','AVAX-USD','LINK-USD', 'LTC-USD','ALGO-USD','BCH-USD','MATIC-USD','XLM-USD',
#         'FIL-USD','TRX-USD','ETC-USD','VET-USD','XTZ-USD','THETA-USD','XMR-USD','EGLD-USD','CRO-USD']

thelen = len(tickers)
thelen
```

```python
price_data = []
for ticker in tickers:
    prices = wb.DataReader(ticker, period='30d')[['Open','Close','Low','High']]
    price_data.append(prices.assign(ticker=ticker)[[ 'Open', 'Close','Low','High','ticker']])

names = pd.concat(price_data)
names.reset_index(inplace=True)
```

```python
df_stocks=names
df_stocks
```

```python
for i in range(2,df_stocks.shape[0]):
    current = df_stocks.iloc[i,:]
    prev = df_stocks.iloc[i-1,:]
    prev_2 = df_stocks.iloc[i-2,:]

    realbody = abs(current['Open'] - current['Close'])
    candle_range = current['High'] - current['Low']

    idx = df_stocks.index[i]
    
#      Bullish swing
    df_stocks.loc[idx,'Bullish swing'] = current['Low'] > prev['Low'] and prev['Low'] < prev_2['Low']

  # Bearish swing
    df_stocks.loc[idx,'Bearish swing'] = current['High'] < prev['High'] and prev['High'] > prev_2['High']

  # Bullish pinbar
    df_stocks.loc[idx,'Bullish pinbar'] = realbody <= candle_range/3 and  min(current['Open'], current['Close']) > (current['High'] + current['Low'])/2 and current['Low'] < prev['Low']

  # Bearish pinbar
    df_stocks.loc[idx,'Bearish pinbar'] = realbody <= candle_range/3 and max(current['Open'] , current['Close']) < (current['High'] + current['Low'])/2 and current['High'] > prev['High']
    
   # Inside bar
    df_stocks.loc[idx,'Inside bar'] = current['High'] < prev['High'] and current['Low'] > prev['Low']
   
   # Outside bar
    df_stocks.loc[idx,'Outside bar'] = current['High'] > prev['High'] and current['Low'] < prev['Low']


   # Bullish engulfing
    df_stocks.loc[idx,'Bullish engulfing'] = current['High'] > prev['High'] and current['Low'] < prev['Low'] and realbody >= 0.8 * candle_range and current['Close'] > current['Open']

   # Bearish engulfing
    df_stocks.loc[idx,'Bearish engulfing'] = current['High'] > prev['High'] and current['Low'] < prev['Low'] and realbody >= 0.8 * candle_range and current['Close'] < current['Open']


df_stocks.fillna(False, inplace=True)
```
